CLEANING THE DATASET

```python
#I dropped the columns i'm not in need of
#Titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True)
Titanic.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], inplace = True )
```

```python
Titanic
```

```python
#Since age is having 177 missing values, I need to fill the missing value in age column so I'm making use of median. From the.describe(), 
# the difference between the min and max in age column is to wide and mode is also not applicable here
median_age = Titanic['Age'].median()
```

```python
median_age
```

```python
#to fill the age column with the median value which is 28.0
Titanic['Age'].fillna(median_age, inplace = True)
```

```python
#To confirm if there are still missing values in the age column
Titanic.isnull().sum() 
```

```python
Titanic
```

```python
Titanic.info()
```

```python
#convert the float number in age to an integer
Titanic['Age'] = Titanic['Age'].astype(int)
```

COMPARING THE SURVIVAL RATES ACROSS INDEPENDENT VARIABLES USING DATA VISUALIZATION

```python
#Survival rate by Pclass using barplot
sns.barplot(x='Pclass', y = 'Survived', data = Titanic)
plt.title ('Survival Rate by Passenger Class')
plt.show()
```

The barplot above indicates that those in higher class which is 1 would likely survive

```python
#Survival rate by sex using barplot
sns.barplot(x='Sex', y = 'Survived', data = Titanic)
plt.title ('Survival Rate by Sex')
plt.show()
```

This bar plot above suggests that females had a higher survival rate than males

```python
#Survival rate by age using lineplot
sns.lineplot(x="Age", y="Survived", data=Titanic)
plt.title("Survival Rate by Age")
plt.show()
```

```python
#Survival rate by fare using histplot
sns.histplot(Titanic[Titanic["Survived"] == 1]["Fare"], bins=30, color="green", kde=True, label="Survived")
sns.histplot(Titanic[Titanic["Survived"] == 0]["Fare"], bins=30, color="red", kde=True, label="Did Not Survive")
plt.legend()
plt.title("Survival Rate by Fare")
plt.show()
```

```python
#Survival rate by SibSp using barplot
sns.barplot(x='SibSp', y = 'Survived', data = Titanic)
plt.title ('Survival Rate by SibSp')
plt.show()
```

```python
#Survival rate by Parch using barplot
sns.barplot(x='Parch', y = 'Survived', data = Titanic)
plt.title ('Survival Rate by Parch')
plt.show()
```

```python
#bring in your label encoder and we are only using it for sex because it is a categorical varaiable. Others are numericals
from sklearn.preprocessing import LabelEncoder
Sex = LabelEncoder()
#fit and transform the sex column
Titanic ['Sex'] = Sex.fit_transform (Titanic['Sex'])
#Titanic['Sex'] = Titanic['Sex'].map({'female':0, 'male':1})
Titanic
```
